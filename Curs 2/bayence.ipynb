```python
from sklearn import datasets
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
```

```python
digit = datasets.load_digits()
breast = datasets.load_breast_cancer()

X_digit = digit.data
y_digit = digit.target
X_breast = breast.data
y_breast = breast.target

X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(digit.data, digit.target, random_state=0, test_size=0.3)
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(breast.data, breast.target, random_state=0, test_size=0.3)

score_bernulli_digit = cross_val_score(BernoulliNB(), X_digit, y_digit).mean()
score_bernulli_breast = cross_val_score(BernoulliNB(), X_breast, y_breast).mean()

score_multinom_digit = cross_val_score(MultinomialNB(), X_digit, y_digit).mean()
score_multinom_breast = cross_val_score(MultinomialNB(), X_breast, y_breast).mean()

score_gaus_digit = cross_val_score(GaussianNB(), X_digit, y_digit).mean()
score_gaus_breast = cross_val_score(GaussianNB(), X_breast, y_breast).mean()

print('Bernoulli: digit = {}, breast = {}'.format(score_bernulli_digit, score_bernulli_breast))
print('Multinomial: digit = {}, breast = {}'.format(score_multinom_digit, score_multinom_breast))
print('Gaussian: digit = {}, breast = {}'.format(score_gaus_digit, score_gaus_breast))
```
